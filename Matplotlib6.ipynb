```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]
y = [0,4,10]
plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("y")
```

```python
fig, axs = plt.subplots(figsize = (6,4))
axs.plot(x,y, linestyle = "dotted", color = "red")
axs.set(title = "graph", xlabel = "x", ylabel = "y")
axs.linecolor = "red"
```

```python
fig, axes = plt.subplots(figsize = (4,6))
x = np.arange(10)
dataframe = pd.Series(np.sin(x),x)
ax = plt.plot(dataframe)
```

```python
cats = ['A','B','C']
values = [4,3,5]
plt.bar(cats, values)
plt.xlabel("cats")
plt.title("breeds")
```

```python
figs, axs = plt.subplots(2,2, figsize = (8,6))
axs[0][1].plot(x,y)
y2 = [4,10,3]
axs[0][0].plot(x,y2)
axs[1][0].bar(cats,values)
```
