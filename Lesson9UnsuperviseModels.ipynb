```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import linear_model
```

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

LR = linear_model.LogisticRegression(solver = 'liblinear', multi_class = 'auto')
LR.fit(X_train, y_train)
LR.score(X_test, y_test)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics.cluster import normalized_mutual_info_score
from sklearn.metrics import adjusted_rand_score
```

```python
from sklearn.cluster import KMeans
Model = KMeans(n_clusters = 3, n_init= "auto").fit(X_train)
pred = Model.fit_predict(X)

#print(f1_score(pred, y))
print(normalized_mutual_info_score(y, pred))
print(adjusted_rand_score(y,pred))
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps = 0.5, min_samples=2).fit(X)
pred = model.fit_predict(X)

print(normalized_mutual_info_score(y, pred))
print(adjusted_rand_score(y,pred))


from sklearn.cluster import HDBSCAN
model = HDBSCAN().fit(X)
pred = model.fit_predict(X)

print(normalized_mutual_info_score(y, pred))
print(adjusted_rand_score(y,pred))
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 3).fit(X)
pred = model.fit_predict(X)

print(normalized_mutual_info_score(y, pred))
print(adjusted_rand_score(y,pred))
```
