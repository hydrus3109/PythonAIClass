```python
dirtiest = pd.DataFrame({
    'Name'     : ['  alice  ', 'Bob', 'Charlie', 'Bob', 'Bob'],
    'Age'      : [25, np.nan, 35, 28,28],
    'City'     : ['new york', 'Los Angeles ', '  chicago', None,None],
    'Fare'     : [200.0, 150.0, 6000, np.nan,np.nan],
    'Survived' : [1, 0, 1, 0,0]
})
```

```python
missing = dirtiest.isna().sum()
print(missing)

cleaner = dirtiest.dropna(subset = ['Age'])
print(cleaner)
```

```python
cleaner = dirtiest
cleaner['City'] = cleaner['City'].fillna('Unknown')
print(cleaner)

cleaner['Age'] = cleaner['Age'].fillna(0)
print(cleaner)
```

```python
cleaner = dirtiest
cleaner['City'] = cleaner['City'].fillna('Unknown')
print(cleaner)

cleaner['Age'] = cleaner['Age'].fillna(cleaner['Age'].mean())
print(cleaner)
```

```python
cleaner = dirtiest
cleaner['City'] = cleaner['City'].fillna('Unknown')
print(cleaner)

cleaner['Age'] = cleaner['Age'].fillna(cleaner['Age'].mean())
print(cleaner)

cleaner['Fare'].median()
cleaner['Fare'] = cleaner['Fare'].fillna(cleaner['Fare'].median())
print(cleaner)
```

```python
cleaner['Survived'] = cleaner['Survived'].astype(bool)
print(cleaner)
cleaner['Age'] = cleaner['Age'].astype(int)
print(cleaner)


```

```python
cleanerer  = cleaner.drop_duplicates()
print(cleanerer)
```

```python
q1, q3 = cleanerer['Fare'].quantile([0.25,0.75])
print(q1,q3)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
print(lower,upper)
cleanerer = cleanerer[(cleanerer['Fare'] >= lower)]
cleanerer = cleanerer[(cleanerer['Fare'] <= upper)]
print(cleanerer)

```

```python
cleanerer['City'] =(cleanerer['City'].str.strip().str.title())
print(cleanerer)
cleanerer['Name'] = (cleanerer['Name'].str.strip())
print(cleanerer)
```
