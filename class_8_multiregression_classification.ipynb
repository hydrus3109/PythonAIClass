```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
from sklearn.metrics import (mean_squared_error, r2_score, accuracy_score, classification_report)
```

ax^5 + bx^4 + cx^3 + d = y

```python
rng = np.random.default_rng(0)
X = rng.uniform(0,2*np.pi, 200).reshape(-1,1)
y = np.sin(X).ravel() + rng.normal(0,0.2, size = len(X))

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)

poly = make_pipeline(PolynomialFeatures(degree = 5), LinearRegression())
poly.fit(X_train, y_train)

y_pred = poly.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
#print("accuracy", accuracy_score(y_test, y_pred) )
xs = np.linspace(0, 2*np.pi, 100).reshape(-1,1)
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, alpha =0.3, label = 'train')
plt.plot(xs, poly.predict(xs), color = "red", label = "fit")
```

```python
from sklearn.datasets import load_wine
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
X, y = load_wine(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
tree = DecisionTreeClassifier(max_depth = 4)
tree.fit(X_train, y_train)
y_pred = tree.predict(y_test)

print(classification_report(y_test, y_pred))
plt.figure(figsize=())
```
